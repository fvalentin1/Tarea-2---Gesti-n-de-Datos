## Ejercicio 3
Genere una función que reciba por parámetro un Diccionario de datos con las notas de 5 alumnos
en un certamen y que devuelva una Serie con las notas de los alumnos aprobados ordenadas de
mayor a menor.

```python
def aprobados(diccionario, nota_aprob = 40):
  alumnos_aprob = []
  notas_aprob =[]

  for x,y in diccionario.items():
    if(diccionario[x] >= nota_aprob):
      alumnos_aprob.append(x)
      notas_aprob.append(y)

  alumnos_aprob_serie = pd.Series(alumnos_aprob)
  notas_aprob_serie = pd.Series(notas_aprob)

  certamenes_dict_aprob = {"Alumno": alumnos_aprob_serie,
                          "Nota": notas_aprob_serie}

  certamenes = pd.DataFrame(certamenes_dict_aprob)
  certamenes = certamenes.set_index("Alumno")

  return certamenes.head(20).sort_values("Nota",ascending=False)

```

```python
notas_certamenes = {"Alvaro": n1,
                    "Benjamin": n2,
                    "Carlos": n3,
                    "David": n4,
                    "Esteban": n5}
                    
aprobados(notas_certamenes)
```

## Ejercicio 8
Los ficheros *emisiones-2016.csv*, *emisiones-2017.csv*, *emisiones-2018.csv* y *emisiones-2019.csv*, contienen datos sobre las emisiones contaminates en la ciudad de Madrid en los años 2016, 2017, 2018 y 2019 respectivamente. Escribir un programa con los siguientes requisitos:

### **a)** Generar un DataFrame con los datos de los cuatro ficheros.

```python
import pandas as pd

ruta_2016 = 'emisiones-2016.csv'
ruta_2017 = 'emisiones-2017.csv'
ruta_2018 = 'emisiones-2018.csv'
ruta_2019 = 'emisiones-2019.csv'

emisiones_2016 = pd.read_csv(ruta_2016, sep = ';')
emisiones_2017 = pd.read_csv(ruta_2017, sep = ';')
emisiones_2018 = pd.read_csv(ruta_2018, sep = ';')
emisiones_2019 = pd.read_csv(ruta_2019, sep = ';')
emisiones = pd.concat([emisiones_2016, emisiones_2017, emisiones_2018, emisiones_2019])

emisiones
```

### **b)** Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION, MAGNITUD, AÑO, MES y las correspondientes a los días D01, D02, etc.



```python
columnas = ['ESTACION', 'MAGNITUD', 'ANO', 'MES']
columnas.extend([col for col in emisiones if col.startswith('D')])
emisiones = emisiones[columnas]

emisiones
```

### **c)** Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los días aparezcan en una única columna.

```python
emisiones = emisiones.melt(id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'], var_name='DIA', value_name='VALOR')

emisiones
```

### **d)** Añadir una columna con la fecha a partir de la concatenación del año, el mes y el día (usar el módulo **datetime**).

```python
import datetime as dt

emisiones['DIA'] = emisiones.DIA.str.strip('D')
emisiones['FECHA'] = emisiones.ANO.apply(str) + '/' + emisiones.MES.apply(str) + '/' + emisiones.DIA.apply(str)
emisiones['FECHA'] = pd.to_datetime(emisiones.FECHA, format='%Y/%m/%d', infer_datetime_format=True, errors='coerce')

emisiones
```

### **e)** Eliminar las filas con fechas no válidas (utilizar la función **isnat** del módulo **numpy**) y ordenar el DataFrame por estaciones, contaminantes y fecha

```python
import numpy as np

emisiones = emisiones.drop(emisiones[np.isnat(emisiones.FECHA)].index)

emisiones.sort_values(['ESTACION', 'MAGNITUD', 'FECHA'])
```

### **f)** Mostrar por pantalla las estaciones y los contaminantes disponibles en el DataFrame.

```python
print('Estaciones disponibles:', emisiones.ESTACION.unique())

print('Contaminantes disponibles:', emisiones.MAGNITUD.unique())
```

### **g)** Crear una función que reciba una estación, un contaminante y un rango de fechas y devuelva una serie con las emisiones del contaminante dado en la estación y rango de fechas dado.

```python
def emisionContaminante(estacion, contaminante, desde, hasta):
  print('Para estación', estacion, 'contaminante', contaminante, 'desde el', desde, 'hasta el', hasta, ':\n')
  return emisiones[(emisiones.ESTACION == estacion) & (emisiones.MAGNITUD == contaminante) & (emisiones.FECHA >= desde) & (emisiones.FECHA <= hasta)].sort_values('FECHA').VALOR
```

```python
emisionContaminante(40, 12, dt.datetime.strptime('2016/05/10', '%Y/%m/%d'), dt.datetime.strptime('2017/05/10', '%Y/%m/%d'))
```

### **j)** Crear una función que reciba una estación y un contaminante y devuelva un resumen descriptivo de las emisiones del contaminante indicado en la estación indicada.

```python
def resumenEmisiones(estacion, contaminante):
  print('Para estación', estacion, 'y contaminante', contaminante, ':\n')
  return emisiones[(emisiones.ESTACION == estacion) & (emisiones.MAGNITUD == contaminante)].VALOR.describe()
```

### **k)** Crear una función que devuelva las emisiones medias mensuales de un contaminante y un año dados para todas las estaciones.

```python
def mediaMensual(contaminante, anio):
  print('Para contaminante', contaminante, 'y año', anio, ':\n')
  return emisiones[(emisiones.MAGNITUD == contaminante) & (emisiones.ANO == anio)].groupby(['ESTACION', 'MES']).VALOR.mean().unstack('MES')
```

```python
mediaMensual(12, 2017)
```

### **l)** Crear una función que reciba una estación de medición y devuelva un DataFrame con las medias mensuales de los distintos tipos de contaminantes.

```python
def mediasMensualesContaminante(estacion):
  print('Para estación', estacion,':\n')
  return emisiones[(emisiones.ESTACION == estacion)].groupby(['MAGNITUD', 'MES']).VALOR.mean().unstack('MES')
```
